# Con quien pasamos mas el tiempo segun nuestra edad?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
file_name='/content/drive/MyDrive/Data_Science/DataSet/time-spent-with-relationships-by-age-us.csv'
df = pd.read_csv(file_name)
```

```python
df.head()
```

```python
df.rename(columns={"Time spent alone, by age of respondent (United States)":"alone",
                   "Time spent with friends, by age of respondent (United States)":"friends",
                   "Time spent with children, by age of respondent (United States)":"children",
                   "Time spent with with parents, siblings and other family, by age of respondent (United States)":"family",
                   "Time spent with partner, by age of respondent (United States)":"partner",
                   "Time spent with coworkers, by age of respondent (United States)":"coworkers"}, inplace=True)
```

```python
df.info()
```

```python
def c_hour(x):
  return round(x /60)
```

```python
df.head()
```

```python
convert_dict = {'alone': int,
                'friends': int,
                'children': int,
                'family' : int,
                'partner': int,
                'coworkers': int
                }

df =  df.astype(convert_dict)
df.head()
```

```python
df.plot('Year', marker="." ,markersize=5, figsize=(15, 10))
plt.ylabel('Tiempo consumido(min al dia)', fontsize=16)
plt.xlabel('Edad', fontsize = 16)
plt.title('Tiempo consumido dedicado por edades (2022)', fontsize=16)
plt.show()
```

#Resultados
El grafico mas arriba presentado, esta basado en encuestas realizadas en Estados Unidos, para determnados grupos de edades en el ano 2022. Este ejercicio se realiza a fin de identificar las tendencias a fin de identificar con que grupo de personas se consumen mayormente el tiempo para las distintas edades.

Se observa que sin importar el rango de edad las personas consumen mayor tiempo solas. No obstante se visualiza que el timpo solo se aumenta a medida que avanzan los anos. Se concluye que las personas de edades mayores a los 60 anos pasan la mayor parte de su tiempo solos.
